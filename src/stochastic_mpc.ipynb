```python
from pyomo.environ import *
from pyomo.opt import SolverFactory
import numpy as np

# Define SMPC model
model = ConcreteModel()

# Define time steps
time_steps = range(len(dates))

# Define variables
model.tilt_angle = Var(time_steps, bounds=(0, 1))  # Control action: 0 (horizontal) to 1 (vertical)
model.GHI_scenario = Var(time_steps, within=NonNegativeReals)

# Setpoint for desired GHI (e.g., desired solar energy level)
ghi_setpoint = 500  # Setpoint in W/m²

# Objective function to minimize deviation from the GHI setpoint across all scenarios
def objective_rule(model):
    return sum((model.GHI_scenario[t] - ghi_setpoint)**2 for t in time_steps) / len(time_steps)

model.objective = Objective(rule=objective_rule, sense=minimize)

# Define the constraints based on each scenario
# We'll use an indexed constraint to set up GHI based on tilt_angle and forecasted GHI for each time step
def ghi_scenario_constraint(model, t, scenario_value):
    # Assuming max GHI occurs at tilt_angle = 0 (horizontal), decrease linearly with tilt_angle
    return model.GHI_scenario[t] == scenario_value * (1 - model.tilt_angle[t])

# Apply the constraints for each time step and scenario
for i, scenario in enumerate(scenario_samples):
    for t in time_steps:
        model.add_component(f"ghi_constraint_scenario_{i}_t_{t}", 
                            Constraint(expr=ghi_scenario_constraint(model, t, scenario[t])))

# Solve the model using a nonlinear solver like IPOPT
solver = SolverFactory('ipopt')
solver.solve(model)

# Extract and display control actions and GHI scenario results
tilt_angles = [model.tilt_angle[t].value for t in time_steps]
GHI_scenarios = [model.GHI_scenario[t].value for t in time_steps]

print("Tilt angles over time:", tilt_angles)
print("GHI scenarios over time:", GHI_scenarios)
```

The warning and the resulting None values for tilt_angles and GHI_scenarios indicate that the solver encountered an issue, likely due to insufficient degrees of freedom in the model. This can happen if there are too many constraints relative to the number of variables or if the problem has inconsistent constraints. Specifically, the message "Too few degrees of freedom" suggests the model is overconstrained or has redundant constraints that cannot all be satisfied simultaneously.

Potential causes and solutions:

1. Overconstrained Model
Cause: The model may have constraints that limit the solution space too tightly, leaving no feasible solution.
Solution: Review the constraints, particularly those that link GHI_scenario and tilt_angle, to ensure they are set up correctly. Simplify or reduce the number of constraints, and test with a single scenario to verify feasibility.
2. Inconsistent or Redundant Constraints
Cause: Redundant constraints, especially if they span multiple scenarios, can overconstrain the model.
Solution: Check that each constraint logically complements others and does not enforce contradictory requirements. Testing with fewer scenarios or constraints can help identify redundancy.
3. Initialization Issues
Cause: Pyomo variables may not be properly initialized, especially when linked across scenarios with multiple constraints.
Solution: Initialize variables explicitly where possible. For example, set initial values for tilt_angle to a feasible guess (like 0.5 for halfway tilt).

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pyomo.environ import *
from pyomo.opt import SolverFactory

# Define a reasonable date range for the scenario
dates = pd.date_range(start="2024-05-01", end="2024-08-31", freq='D')
time_steps = range(len(dates))

# Generate synthetic GHI data for illustration (replace with actual data for real analysis)
np.random.seed(42)
forecasted_ghi = np.random.normal(450, 30, len(dates))  # Forecasted GHI around 450 W/m² with 30 W/m² std deviation
real_ghi = forecasted_ghi + np.random.normal(0, 20, len(dates))  # Real GHI deviates by additional 20 W/m²

# Create dataframes for merging
historical_forecasted = pd.DataFrame({'date': dates, 'ghi_forecasted': forecasted_ghi})
historical_real = pd.DataFrame({'date': dates, 'ghi_real': real_ghi})

# Merge dataframes to align dates
merged_data = pd.merge(historical_real, historical_forecasted, on='date')
merged_data['ghi_difference'] = merged_data['ghi_real'] - merged_data['ghi_forecasted']

# Calculate the mean and standard deviation of the difference
mean_diff = merged_data['ghi_difference'].mean()
std_diff = merged_data['ghi_difference'].std()

# Generate scenarios for SMPC by adding random noise based on calculated mean and std deviation
num_scenarios = 5  # Reduced for simplicity in testing feasibility
scenario_samples = [
    merged_data['ghi_forecasted'] + np.random.normal(mean_diff, std_diff, len(merged_data))
    for _ in range(num_scenarios)
]

# Initialize the Pyomo model
model = ConcreteModel()

# Define model variables with more realistic bounds and initial values
model.tilt_angle = Var(time_steps, bounds=(0, 1), initialize=0.5)  # Tilt angle bounds from 0 to 1
model.GHI_scenario = Var(time_steps, within=NonNegativeReals, initialize=450)  # Starting point for GHI scenario

# Setpoint for desired GHI to ensure it's within a feasible range
ghi_setpoint = 450  # Setpoint in W/m², realistic target for summer days

# Define the objective function to minimize deviation from the GHI setpoint across all scenarios
def objective_rule(model):
    return sum((model.GHI_scenario[t] - ghi_setpoint)**2 for t in time_steps) / len(time_steps)

model.objective = Objective(rule=objective_rule, sense=minimize)

# Define a constraint function to model GHI response based on tilt angle and scenarios
def ghi_scenario_constraint(model, t, scenario_value):
    max_ghi = scenario_value  # Max GHI from scenario data at horizontal tilt
    return model.GHI_scenario[t] == max_ghi * (1 - model.tilt_angle[t])

# Add constraints for each time step and scenario
for i, scenario in enumerate(scenario_samples):
    for t in time_steps:
        model.add_component(f"ghi_constraint_scenario_{i}_t_{t}", 
                            Constraint(expr=ghi_scenario_constraint(model, t, scenario[t])))

# Solve the model using the IPOPT solver
solver = SolverFactory('ipopt')
result = solver.solve(model, tee=True)

# Check for model feasibility and display outputs if successful
if (result.solver.termination_condition == TerminationCondition.optimal):
    tilt_angles = [model.tilt_angle[t].value for t in time_steps]
    GHI_scenarios = [model.GHI_scenario[t].value for t in time_steps]

    print("Tilt angles over time:", tilt_angles)
    print("GHI scenarios over time:", GHI_scenarios)
else:
    print("Solver did not find an optimal solution. Please check constraints and initial values.")
```

GHI Setpoint: Set to 450 W/m², which is realistic for sunny days and within reach of typical daily solar irradiance.
Scenario Generation: Reduced the number of scenarios to five for easier troubleshooting and lower complexity.
Constraints on Tilt Angle: Set tilt angle bounds between 0 and 1 (horizontal to vertical).
Initialization of Variables: Initialized tilt_angle to 0.5 and GHI_scenario to 450 to help the solver find a feasible path.
# Expected Output (results for control action and GHI estimates over time)
Tilt Angles: Values between 0 and 1, adjusting to meet the GHI setpoint under each scenario.
GHI Scenarios: Values close to the setpoint, as the model attempts to minimize deviation across scenarios.

1. Plotting the Mean and Standard Deviation of Forecast Errors Over Time

```python
import matplotlib.pyplot as plt
import numpy as np

# Transpose the scenario samples so each entry in `scenario_samples_by_time` represents a time step
scenario_samples_by_time = np.array(scenario_samples).T

# Calculate the mean and standard deviation of deviations from the setpoint at each time step
mean_deviation = [scenario_samples_by_time[t].mean() - ghi_setpoint for t in time_steps]
std_deviation = [scenario_samples_by_time[t].std() for t in time_steps]

# Plot the mean and standard deviation over time
plt.figure(figsize=(12, 6))

# Plot mean deviation over time
plt.subplot(2, 1, 1)
plt.plot(dates, mean_deviation, label='Mean Deviation from Setpoint', color='blue')
plt.axhline(0, color='gray', linestyle='--', alpha=0.7)  # Desired setpoint line
plt.xlabel('Date')
plt.ylabel('Mean Deviation (W/m²)')
plt.title('Mean Deviation from GHI Setpoint Over Time')
plt.legend()

# Plot standard deviation (uncertainty) over time
plt.subplot(2, 1, 2)
plt.plot(dates, std_deviation, label='Standard Deviation (Uncertainty)', color='red')
plt.xlabel('Date')
plt.ylabel('Standard Deviation (W/m²)')
plt.title('Uncertainty (Standard Deviation of Deviation) Over Time')
plt.legend()

plt.tight_layout()
plt.show()
```

scenario_samples_by_time: This array organizes scenario_samples by time steps, where each element contains values from all scenarios at a specific time step.
mean_deviation and std_deviation: Calculated across scenarios for each time step to illustrate how mean deviation and uncertainty evolve over time.

Mean Deviation: This plot shows how the average deviation from the setpoint evolves over time. Ideally, it should approach zero if the SMPC is working to track the setpoint closely.

Standard Deviation (Uncertainty): If uncertainty is converging, the standard deviation should decrease over time, indicating reduced forecast error and increased model stability.

2. Plotting Confidence Intervals for Scenario Deviation Over Time

```python
# Transpose scenario_samples so each sub-list represents values at a single time step across scenarios
scenario_samples_transposed = np.array(scenario_samples).T  # Each row is now a time step, each column is a scenario

# Calculate the 10th, 50th (median), and 90th percentiles of scenario deviations from setpoint at each time step
percentile_10 = [np.percentile(scenario_samples_transposed[t] - ghi_setpoint, 10) for t in time_steps]
percentile_50 = [np.percentile(scenario_samples_transposed[t] - ghi_setpoint, 50) for t in time_steps]
percentile_90 = [np.percentile(scenario_samples_transposed[t] - ghi_setpoint, 90) for t in time_steps]

# Plot the confidence intervals over time
plt.figure(figsize=(10, 6))
plt.fill_between(dates, percentile_10, percentile_90, color='skyblue', alpha=0.4, label='10-90th Percentile Range')
plt.plot(dates, percentile_50, color='blue', label='Median Deviation')
plt.axhline(0, color='gray', linestyle='--', alpha=0.7)  # Desired setpoint line
plt.xlabel('Date')
plt.ylabel('Deviation from Setpoint (W/m²)')
plt.title('Confidence Interval of Scenario Deviations Over Time')
plt.legend()
plt.show()
```

scenario_samples_transposed: By transposing, each row now represents values at a specific time step across all scenarios. This allows us to calculate percentiles correctly for each time step.
Percentile Calculation: The percentiles are calculated for each time step based on deviations from the setpoint.

The shaded area represents the range between the 10th and 90th percentiles of the scenario deviations.
Narrowing of the Confidence Interval: If the shaded area narrows as time progresses, it indicates that uncertainty is decreasing, showing convergence.


The cumulative distribution of forecast errors plot provides insight into the spread and concentration of forecast errors (or deviations from the setpoint) by showing the cumulative probability of the deviations up to a given point.

Interpretation of the Cumulative Distribution Plot:

X-axis (Deviation from Setpoint): This axis represents the difference between the forecasted or scenario-based GHI and the desired GHI setpoint. Values closer to zero indicate small deviations, meaning the forecast was close to the target.

Y-axis (Cumulative Probability): This axis shows the cumulative probability of deviations, which increases as we move along the x-axis. It represents the proportion of deviations that are less than or equal to a given deviation value on the x-axis.

Slope and Spread:
Sharp Rise Near Zero: If the cumulative distribution has a sharp rise near zero, this indicates that most of the deviations are small, suggesting that the forecast or scenario-based predictions are accurate. A steep curve centered around zero deviation indicates higher forecast accuracy and less uncertainty.

Flat Slope or Spread Out Curve: If the cumulative distribution is more spread out, it indicates greater variation in the forecast errors. A flatter curve means there are larger deviations from the target, indicating more forecast uncertainty and less accurate scenario predictions.

Overall Distribution Shape:
If the distribution is symmetric and steep around zero, it suggests that the forecast or scenario predictions are generally close to the target with minor errors.
If the distribution is skewed or shows a long tail (e.g., on the positive side), it may suggest a consistent bias in the forecast (overestimation or underestimation).

conclusion:

If the cumulative distribution plot shows that around 90% of the deviations fall within a small range (e.g., ±20 W/m²), you can conclude that uncertainty is converging, as the majority of forecasts or scenario deviations are close to the desired GHI setpoint. Conversely, a wide spread would indicate high variability, suggesting that uncertainty is still significant and forecasts are less reliable.
