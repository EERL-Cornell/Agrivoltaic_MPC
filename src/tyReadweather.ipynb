1. Inputs:
GHI values and solar zenith angles (theta_z).
2. Clearness Index kt:
The clearness index is calculated using GHI, the extraterrestrial irradiance E0, and the cosine of the zenith angle.
3. Diffuse Fraction fd:
The Erbs model is used to compute the diffuse fraction based on the clearness index.
4. DHI:
The Diffuse Horizontal Irradiance is calculated as DHI=GHI×fd.
5. DNI:
Finally, DNI is calculated using the known relationship.
